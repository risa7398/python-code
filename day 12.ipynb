```python
#tree traverse

class node:
    def __init__(self,value):
        self.data=value
        self.tag=None
        self.left=None
        self.right=None

def preorder(root):
    if root==None:
        return
    print(root.data,end=" ")
    preorder(root.left)
    preorder(root.right)

def postorder(root):
    if root==None:
        return
    preorder(root.left)
    preorder(root.right)
    print(root.data,end=" ")

def inorder(root):
    if root==None:
        return
    inorder(root.left)
    print(root.data,end=" ")
    inorder(root.right)

def levelorder(root):
    q=[root,None]
    while q:
        curr=q.pop(0)
        if curr==None:
            print()
            if q:
                q.append(curr)
        else:
            print(curr.data,end=" ")
            if curr.left!=None:
                q.append(curr.left)
            if curr.right!=None:
                q.append(curr.right)
    
root=node(1)
root.left=node(2)
root.right=node(3)

root.left.left=node(4)
root.left.right=node(5)

root.right.left=node(6)
root.right.right=node(7)

print("\npreorder:")
preorder(root)

print("\npostorder:")
postorder(root)

print("\ninorder:")
inorder(root)

print("\nlevelorder:")
levelorder(root)
```
